```python
import torch
from langchain_huggingface import HuggingFacePipeline
from transformers import pipeline
from langchain_core.prompts import PromptTemplate

model = pipeline("text2text-generation", model="google/flan-t5-large", device="cpu", max_length=1024, truncation = True)
llm = HuggingFacePipeline(pipeline=model)
template = PromptTemplate.from_template("Explain {topic} in detail to a {age} year old.")
chain = template | llm
age = input("Enter the age of the person: ")
topic = input("Enter the topic you want to learn about: ")
prompt = chain.invoke({"topic": topic, "age": age})
print(prompt)
```

```python
import torch
from langchain_huggingface import HuggingFacePipeline
from transformers import pipeline
from langchain_core.prompts import PromptTemplate
from transformers.utils.logging import set_verbosity_error
set_verbosity_error()
summarisation = pipeline("summarization", model="facebook/bart-large-cnn", device="cpu", max_length=512, truncation=True)
summariser = HuggingFacePipeline(pipeline=summarisation)
refinement = pipeline("summarization", model="facebook/bart-large", device="cpu", max_length=1024, truncation=True)
refiner = HuggingFacePipeline(pipeline=refinement)
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2", device="cpu", max_length=512, truncation=True)
qa = HuggingFacePipeline(pipeline=qa_pipeline)
summary_template = PromptTemplate.from_template("Summarise the following text, {text} in {length} words.")
text_to_summarise = input("Enter the text you want to summarise: ")
length = input("Enter the length of the summary in words: ")
summary_chain = summary_template | summariser | refiner
summary = summary_chain.invoke({"text": text_to_summarise, "length": length})
print("Summary: ", summary)

while True:
    question = input("Enter your question (or type 'exit' to quit): ")
    if question.lower() == 'exit':
        break
    
    result = qa_pipeline(question=question, context=summary)
    
    print(f"Answer: {result['answer']}")
    print(f"Confidence Score: {round(result['score'], 4)}")
```
