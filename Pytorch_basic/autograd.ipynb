```python
import torch
```

## Manual

```python
x = torch.tensor(6.7)
y = torch.tensor(0)
w = torch.tensor(1.0)
b = torch.tensor(0.0)
```

```python
def binary_cross_entropy(prediction, target):
    epsilon = 1e-8
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return -(target*torch.log(prediction) + (1-target)*torch.log(1-prediction))
```

```python
z = w * x + b
y_pred = torch.sigmoid(z)
loss = binary_cross_entropy(y_pred, y)
```

```python
loss
```

```python
dloss_dy_pred = (y_pred - y) / (y_pred*(1-y_pred))
dloss_dy_pred_dz = y_pred * (1 - y_pred)
dz_dw = x
dz_db = 1.0
```

```python
dL_dw = dloss_dy_pred * dloss_dy_pred_dz * dz_dw
dL_db = dloss_dy_pred * dloss_dy_pred_dz * dz_db
```

```python
print(f"Manual gradient calc of loss wrt w:{dL_dw}")
print(f"Manual gradient calc of loss wrt b:{dL_db}")
```

## Autograd

```python
x = torch.tensor(6.7)
y = torch.tensor(0)
w = torch.tensor(1.0, requires_grad = True)
b = torch.tensor(0.0, requires_grad = True)
```

```python
z = w * x + b
y_pred = torch.sigmoid(z)
loss = binary_cross_entropy(y_pred, y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```
